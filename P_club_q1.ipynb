```python
def loss(y_pred,y_true):

  n=len(y_pred)
  for i in range (0,n):
    sum=(y_true[i]-y_pred[i])**2

  mse=sum/n

  return mse
```

```python
def y(x,a,b):
  y_pred=[]
  m,n = x.shape
  for i in range (0,m):
    sum=0
    for j in range (0,n):
      sum+=x[i][j]*a[i]
    y_pred[i]=sum+b


  assert(y_pred.shape == (m,))
  return y_pred
```

```python
import sympy as sp
def gradient(x,a,b,y_true):
  m,n = x.shape
  yp = y(x,a,b)
  loss=yp-y_true

  da = sp.diff(loss,a)              # write code to calculate derivative of loss with respect to a
  db = sp.diff(loss,b)              # write code to calculate derivative of loss with respect to b

  assert(da.shape ==(n,))
  return (da,db)
```

```python
def gradient_descent(x,y_true,learning_rate=0.01,epochs = 10):
  """
  This function perfroms gradient descent and minimizes loss
  input:
  x: [array] the feature vector of shape (m,n)
  y_true: [array] ground truth of shape (m,)

  output:
  loss: [array] of size (epochs,)
  weights: [tuple] (a,b)
  """
  m,n = x.shape
  loss_mse = []                                 # initialize empty list to store loss
  a = []                                       # initialize a- weights and b- bias
  b = []

  for i in range(epochs):
    # calculate derivative using gradient() function
    derv=gradient(x,a,b,y_true)
    # apply gradient descent now to update a and b
    a=a-learning_rate*derv[0]
    b=b-learning_rate*derv[1]
    y_pred=y(x,a,b)


    l_mse = loss(y_pred,y_true)                               # calculate loss at this point
    loss_mse.append(l_mse)

    print("Epoch ",i+1," Completed!","loss = ",l_mse)

  print("Training completed!!")

  assert(a.shape==(n,))

  return (loss_mse,a,b)
```
